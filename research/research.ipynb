```python
def get_product(x: int, y: int) -> int:
    return x * y 
print(get_product(2, 3))
print(get_product(2, "88"))
```

```python
from ensure import ensure_annotations

@ensure_annotations
def get_productt(x: int, y: int) -> int:
    return x * y
print(get_productt(2, 3))
print(get_productt(2, "88"))
```
